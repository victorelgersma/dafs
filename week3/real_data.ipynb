```python
eu_countries = set([
    "Austria","Belgium","Bulgaria","Croatia","Cyprus","Czechia",
    "Denmark","Estonia","Finland","France","Germany","Greece",
    "Hungary","Ireland","Italy","Latvia","Lithuania","Luxembourg",
    "Malta","Netherlands","Poland","Portugal","Romania","Slovakia",
    "Slovenia","Spain","Sweden", "United Kingdom"
])
```

```python
# check if we have all countries: 

EU_DATA = df[df["country"].isin(eu_countries)].copy()

present = set(EU_DATA["country"].unique())
required = set(eu_countries)

not_present = required - present

if not not_present:
    print("All countries present")
else:
    print(f"missing: {not_present}")
```

```python

print(EU_DATA.describe())

print(EU_DATA.head(15))


columns_share_energy = ["solar_share_energy", "wind_share_energy", "biofuel_share_energy", "nuclear_share_energy", "hydro_share_energy", "other_renewables_share_energy"]

EU_DATA[columns_share_energy][EU_DATA["country"] == "Netherlands"]
```

```python
EU_DATA["renewable_percentage"] = EU_DATA[columns_share_energy].sum(axis=1, min_count=1) # row-wise sum

```

```python
#Check if this worked
EU_DATA[["country", "year", "renewable_percentage"]][EU_DATA["year"] == 2022].sort_values(by="renewable_percentage", ascending=False).set_index("country")
```

```python
EU_DATA["renewable_category"] = pd.cut(
    EU_DATA["renewable_percentage"],
    bins=[0, 20, 50, 100],             # thresholds
    labels=["Low Renewable", "Medium Renewable", "High Renewable"],
    include_lowest=True # What does this do? 
)

# https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.cut.html
```

```python
EU_DATA[["country", "year", "renewable_percentage", "renewable_category"]][EU_DATA["year"] == 2023].sort_values(by="renewable_percentage", ascending=False)
```

```python
# visualisation (not part of exercise but interesting)
import matplotlib.pyplot as plt 

fig, ax = plt.subplots() 

highlight = ["Netherlands", "Sweden", "France", "Bulgaria"]


for country in eu_countries:
    y = EU_DATA[EU_DATA["country"] == country][["renewable_percentage"]]
    x = EU_DATA[EU_DATA["country"] == country][["year"]]

    if country in highlight:
        ax.plot(x,y, label=country, linewidth = 2.5)
    else :
        ax.plot(x,y, color= 'lightgrey', linewidth = 1)



ax.legend(title="Country")
ax.set_xlabel("Year")
ax.set_ylabel("Share of Renewables in Energy Mix")
```
