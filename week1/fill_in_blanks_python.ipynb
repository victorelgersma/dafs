## Exercise 4 — Concatenate text
**Goal:** Build strings with space and with an underscore.

```python
# We have a first and last name, we want to combine them into one string
# 1) We want a string in which the names are separeted with a space
# 2) We want a string in which the names are separeted with a "_"

first = "Freya"
last  = "Crescent"
 
full1 = first + " " + last          
full2 = first + "_" + last
print(full1)
print(full2)

check(full1 == "Freya Crescent" and full2 == "Freya_Crescent")
```

## Exercise 6 — Minutes to hours:minutes
**Goal:** Use integer division `//` and modulo `%`.
**Goal:** Transform a number of minutes into a string tahat gives the number of hours and minutes (i.e 4 hours 15 minutes instead of 255 minutes). You will need an operator for the division and to compute the remainder (also called modulo).  Find these functions online. 

```python
minutes = 135
hours = minutes // 60   
left  = minutes % 60  

msg = f"{hours} hour(s) and {left} minute(s)"
print(msg)

check(hours == 2 and left == 15)
```

## Exercise 7 — Logic operators
**Goal:** Create a script that checks the age to determine if someone is eligible for a student discount
The discount is valid if the person is either a student or age <= 18

```python
age = 19
student = True

is_adult = age >= 18               
discount_eligible = student or not is_adult   

print(is_adult)
print(discount_eligible)

check(is_adult is True and discount_eligible in (True, False))
```
